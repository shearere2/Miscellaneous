```python
def safe(levels: list) -> bool: # Check if a list is safe
  curr = 0
  for i in range(len(levels)-1):
    sign = copysign(1, int(levels[0]) - int(levels[1]))
    if copysign(1, int(levels[i]) - int(levels[i+1])) != sign:break
    if abs(int(levels[i]) - int(levels[i+1])) not in [1, 2, 3]: break
    else:
      curr += 1
      if curr == len(levels) - 1: return True
```
