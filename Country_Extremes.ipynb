```python
import pandas as pd
df = pd.read_csv('country_extremes.csv')
```

```python
def floatify(c):
  c = c.replace(',', '')
  if c == 'sea level':
    return 0.0
  if c[0] == '−':
    return -float(c[1:-2])
  return float(c[:-2])
```

```python
df['Minimum elevation'] = df['Minimum elevation'].apply(floatify)
df['Maximum elevation'] = df['Maximum elevation'].apply(floatify)
df['Elevation span'] = df['Elevation span'].apply(floatify)
df = df.rename(columns={'Country or region':'region', 'Continent':'continent',
                        'Highest point':'high_point', 'Maximum elevation':'max_elev_meters',
                        'Lowest point':'low_point', 'Minimum elevation':'min_elev_meters',
                        'Elevation span':'elev_span_meters'})
```

```python
countries = pd.read_csv('countries of the world.csv')
```

```python
countries['area_sq_km'] = countries['Area (sq. mi.)'].apply(lambda x: float(x)*2.59)
```

```python
countries = countries.rename(columns={'Country':'region'})
countries['region'] = countries['region'].apply(lambda x: x.strip())
df['region'] = df['region'].apply(lambda x: x.strip('\xa0'))
df = df.merge(countries[['region', 'area_sq_km', 'Population']], on='region')
```

```python
df['hilly_index'] = df['elev_span_meters']/df['area_sq_km']
```

```python
df.sort_values(by='hilly_index')
```
